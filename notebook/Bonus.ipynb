```python
dataset = spark.read.csv('abfss://disco1@storageguidatalake.dfs.core.windows.net/data/dataset.csv/part-000??-69dc1255-b664-46f3-b53c-0cc7e120de8a-c000.csv', sep=',',header=True)
```

```python
from pyspark.sql import functions as f
```

## Bônus 1

Utilizando a coluna do dia da semana e filtrando pra sábado, domingo e segunda.

```python
bonus_1 = dataset\
  .select('*')\
  .where((dataset.dia_da_semana == 7) | (dataset.dia_da_semana == 1) | (dataset.dia_da_semana == 2))
```

```python
bonus_1.agg({'tempo_viagem': 'avg'}).show()
```

## Bônus 2

```python
bonus_2 = dataset\
  .select('*', f.year('Data').alias('Ano'))\
  .where('Ano==2010')\
  .limit(500)
```

```python
df_mapa = bonus_2.select('dropoff_latitude', 'dropoff_longitude', 'pickup_latitude', 'pickup_longitude').toPandas()
```

```python
import folium
```

```python
m = folium.Map(location=[40.730610, -73.935242], zoom_start=15)
m
```

```python
for k, v in df_mapa.iterrows():
  folium.Marker(
    location=[v['dropoff_latitude'], v['dropoff_longitude']],
    popup="Mt. Hood Meadows",
    icon=folium.Icon(icon="cloud", color='red'),
  ).add_to(m)
  folium.Marker(
    location=[v['pickup_latitude'], v['pickup_longitude']],
    popup="Mt. Hood Meadows",
    icon=folium.Icon(icon="cloud", color='green'),
  ).add_to(m)
```
